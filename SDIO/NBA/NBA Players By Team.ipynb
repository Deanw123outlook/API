# NBA Players By Team

SDIO Subscription has been removed for data privacy purposes.

```python
import json # Python built-in package json for encoding and decoding JSON data.
import requests # library for working with HTTP requests in Python
import pandas as pd
import numpy as np
import time
```

```python
teams = ['CHI','DET','CLE','BOS','SAC'] # Sample team names
```

```python
# build 9 empty list to extract data from API runs
playerID,Status,TeamID,Team,PositionCategory,Position,FirstName,LastName,BirthDate = ([] for i in range(9))
```

```python
for team in teams:
    #
    print("------------------------------------------------------------------------------------------------------------------")
    
    # build stings using list of dates  
    string = 'https://api.sportsdata.io/v3/nba/scores/json/PlayersBasic/{}?key=x' # {} acts as string placeholder
    string = string.format(team) # format() method formats the specified date(s) and insert them inside the string's placeholder.
    
    # build URL parameters
    url = (string)
    response = requests.get(url) # get requests
    print(url)

    # Convert api get request body to JSON object
    data = response.json() # convert to JSON 
    print("Request-Status-Code : {}".format(response.status_code)) # get request status code
    
    for player in data:
        print(player['PlayerID'])
        playerID.append(player['PlayerID'])
        print(player['Status'])
        Status.append(player['Status'])
        print(player['TeamID'])
        TeamID.append(player['TeamID'])
        print(player['Team'])
        Team.append(player['Team'])
        print(player['PositionCategory'])
        PositionCategory.append(player['PositionCategory'])
        print(player['Position'])
        Position.append(player['Position'])
        print(player['FirstName'])
        FirstName.append(player['FirstName'])
        print(player['LastName'])
        LastName.append(player['LastName'])
        print(player['BirthDate'])
        BirthDate.append(player['BirthDate'])
        print()
        
    #
    time.sleep(2)
        
```

```python
#Create DataFrame including all above features and randomly generated data
df = pd.DataFrame ({'PlayerID':playerID,'Status':Status,'TeamID':TeamID,'Team': Team,
                   'PositionCategory':PositionCategory,'Position':Position,'FirstName':FirstName,'LastName':LastName,
                   'BirthDate':BirthDate})

#create a copy of original dataframe ...(good practice)
original_dataframe = df.copy(deep = True)
df
```
