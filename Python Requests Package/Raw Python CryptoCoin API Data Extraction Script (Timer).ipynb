Endpoint Parameters

```python
url = "https://api.livecoinwatch.com/coins/list" # url endpoint

payload = json.dumps({ # json.dumps() convert Python object into a JSON string
  "currency": "USD",
  "sort": "rank",
  "order": "ascending",
  "offset": 0,
  "limit": 100, # 20 - 100 --->>> Coin Watch API
  "meta": False
})

# headers
headers = {
  'content-type': 'application/json',
  'x-api-key': '98b16066-5b97-48bd-aa0c-66fae75a6db7' # Subscritpion Key
}
```

Build Dataframes & Data Structures to employ in our API data extraction model

```python
cd, r, v, cp = ([] for i in range(4))
time_stamp = [] # empty list

#Create DataFrame from our models data collected
base_df = pd.DataFrame ({'Coin':cd,'Rate':r,'Volume':v,'Cap':cp})
base_metadata_df = pd.DataFrame ({'API-Request-Timestamp':time_stamp})
```

```python
iteration = 0 # declare variable to increment within our model

# API DATA EXTRACTION MODEL
for test in range(1,200,1):
    iteration += 1 # incremenrt (manual iteration tracker)
    print("Model-Iteration:",iteration)
    #
    response = requests.request("POST", url, headers=headers, data=payload)
    
    # timestamp
    dt_request = datetime.now() # Getting the current date and time request executed/processes (time will be close to time server was hit for request)
    timestamp = datetime.timestamp(dt_request) # Getting the timestamp (time will be close to time server was hit for request)
    time_stamp.append(dt_request)
    
    #
    metadata = pd.DataFrame({'API-Request-Timestamp':time_stamp})
    meta_dataset = pd.merge(base_metadata_df, metadata, how="outer")
    
    #
    data = response.json()
    
    # API Data Extraction
    for item in data:
        #
        cd.append(item['code'])
        r.append(item['rate'])
        v.append(item['volume'])
        cp.append(item['cap'])
        
        #
        df = pd.DataFrame({'Coin':cd,'Rate':r,'Volume':v,'Cap':cp})
        dataset = pd.merge(base_df, df, how="outer")
    
    print("--timer-sleep--function--start--")
    print(datetime.now())
    time.sleep(30) # sleep function takes a float argument representing the number of seconds that the program will pause for
    print(datetime.now())
    print("--timer-sleep--function--end--")
    
```
